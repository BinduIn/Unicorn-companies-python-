```python
df['Age'] = (datetime.now() - df['Date Joined']).apply(lambda x: x.days)
```

```python
df['Daily Value ($M)'] = (df['Valuation ($B)'] / df['Age']) * 1000
```

```python
df.head()
```

```python
sns.scatterplot(data=df, x = 'Age', y = 'Daily Value ($M)', hue='Industry', size='Valuation ($B)', palette = 'Spectral')
```

```python
bins = [0, 5, 101, 500]
group_names = ['walk','trot','gallop'] #sticking with the unicorn theme
df['Speed'] = pd.cut(df['Daily Value ($M)'], bins, labels=group_names)
```

```python
df.head(5)
```

```python
sns.scatterplot(data=df, x = 'Age', y = 'Daily Value ($M)', hue='Speed', palette = 'Spectral')
```

```python
def industry_leaders(industry):
    print(industry)
    c = df.loc[df.Industry==industry].head(10).sort_values(by = 'valuation',ascending= False)
    li_gph = px.bar(c,x ='Company',y='valuation',template='plotly_dark')
    li_gph.show()
```

```python
industry_leaders('Edtech')
```

```python
def industry_leader_along_with_country(industry):
    print(industry)
    c = df.loc[df.Industry==industry].head(10).sort_values(by = 'valuation',ascending= False)
    li_gph = px.bar(c,x ='Company',y='valuation',template='plotly_dark',color='Country')
    li_gph.show()
```

```python
industry_leader_along_with_country('Edtech')
```
